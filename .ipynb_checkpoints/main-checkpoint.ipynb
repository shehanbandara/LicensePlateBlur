```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
carPlate = cv2.imread('carPlate.jpg')
```

```python
def display(image):
    figure = plt.figure(figsize=(10,8))
    ax = figure.add_subplot(111)
    newImage = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    ax.imshow(newImage)
```

```python
display(carPlate)
```

```python
plateCascade = cv2.CascadeClassifier('haarcascaderussianPlateNumber.xml')
```

```python
def detectPlate(image):
    plateImage = image.copy()
    plateRectangle = plateCascade.detectMultiScale(plateImage, scaleFactor=1.3, minNeighbors=3)
    for (x,y,w,h) in plateRectangle:
        cv2.rectangle(plateImage, (x,y), (x+w,y+h), (0,0,255), 4)
    return plateImage
```

```python
result = detectPlate(carPlate)
display(result)
```

```python
def detectAndBlurPlate(image):
    plateImage = image.copy()
    regionOfInterest = image.copy()
    plateRectangle = plateCascade.detectMultiScale(plateImage, scaleFactor=1.3, minNeighbors=3)
    for (x, y, w, h) in plateRectangle:
        regionOfInterest = regionOfInterest[y:y+h, x:x+w]
        blurredRegionOfInterest = cv2.medianBlur(regionOfInterest, 7)
        plateImage[y:y+h, x:x+w] = blurredRegionOfInterest
    return plateImage
```

```python
result = detectAndBlurPlate(carPlate)
display(result)
```
